```python

import pandas as pd
data1={
    "Name":['Ram','Shyam','jff'],
    "Age":[10,20,30],
    "City":['Nagpur','Mumbai','Delhi'],
    "Salary":[10000,20000,30000]
}
# print(data)
df1=pd.DataFrame(data1)
print(df1)
print()

print("How to add, remove,update in the column")
print()

print("Adding Data--------------------------------------")
print("there is two ways, direct assign, usingn insert()")
print()

df1["Bonus"]=df1["Salary"]*0.1
print(df1)
print()

print("Using 2 method insert()")
df1.insert(0,"Emp_ID",[10,20,30])
print(df1)
print()
print("Update data---------------------------------------")
print()

df1.loc[0,"Salary"]=15000        #loc[row index,col_name]=new_value
print(df1)

print()
print("If we want to update multiple values")

df1["Salary"]=df1["Salary"]*1.05       
print(df1)
print()

df1.loc[df1["Name"] == "Shyam", "Salary"] = 25000
print(df1)

print()
df1.loc[df1["Name"].isin(["Ram", "Shyam"]), "Salary"] = 50000
print(df1)

print()
df1.loc[[0,1],"Salary"]=[30000,40000]
print(df1)

print()
print("Removing columns----------------------------------")

print("not Remove from main(raw) data if [inplace is false]")
df2=df1.drop(columns=["Bonus",'Age'],inplace=False)
print(df2)
print()

print("Remove from main data if [inplace is true]")
df1.drop(columns=["Bonus"],inplace=True)    
print(df1)
print()

print("Remove multi column")
df1.drop(columns=["Name","Age"],inplace=True) 
print(df1)


```

```python
print("How to handling missing data------------------")
'''
NAN (not a number)
None(for object data types)

is NULL()
True-NaN is missing
False- value is present

'''


import pandas as pd
import numpy as np
dataa={
    "Name":['Ram','None','jff',None],
    "Age":[10,0,30,None],
    "City":['Nagpur',None,'Delhi',0],
    "Salary":[10000,None,30000,None]
}
# print(data)
dff=pd.DataFrame(dataa)
print(dff)
print()
print("HOw to Check if Null is present")
print(dff.isnull())

print()
print("How many missing data present")
print(dff.isnull().sum())

print()
print("How to drop missing data    or    fill the missing data")
print()
print(dff)
print()

dff.dropna(inplace=True)
print(dff)

print()
print("How to fill nan values..............")

dataa1={
    "Name":['Ram','None','jff',None],
    "Age":[10,0,30,None],
    "City":['Nagpur',None,'Delhi',0],
    "Salary":[10000,None,30000,None]
}
dff2=pd.DataFrame(dataa1)
print(dff2)
print()
print("fill as a default value")
dff2.fillna(0,inplace=True)   #fillna(value,inplace=true,false)
print(dff2)
print()
dff3=pd.DataFrame(dataa1)
print(dff3)
print("Add some vakue based on mean")
dff3["Age"] = dff3["Age"].fillna(dff3["Age"].mean())
print(dff3)
print()
dff3["Salary"] = dff3["Salary"].fillna(dff3["Salary"].mean())
print(dff3)

print()
print("n pandas, interpolation is a method used to fill missing values (NaN) using existing data patterns instead of fixed values like 0 or mean.")
print()
df3 = pd.DataFrame({
    "Age": [10, None, 30, None, 50]
})

df3["Age"] = df3["Age"].interpolate()
print(df3)

```

```python
print("n pandas, interpolation is a method used to fill missing values (NaN) using existing data patterns instead of fixed values like 0 or mean.")
print()
df3 = pd.DataFrame({
    "Age": [10, None, 30, None, 50]
})

df3["Age"] = df3["Age"].interpolate()
print(df3)

print()
print("Linear Mrthod")
df3.interpolate(method='linear', axis=0, inplace=False)
print(df3)
print()
print("Forward fill")
df4 = pd.DataFrame({
    "Age": [10, None, 30]
})

print(df4.ffill())
print()
print("Backward fill")
df4.bfill()
print(df4)
```

```python
print("Sorting and aggegration")
print()
print('''
📘 Sorting and Aggregation in pandas

In pandas, these are two core operations:

Sorting → arranging data (ascending/descending)
Aggregation → summarizing data (sum, mean, count, etc.)

df.sort_values(by="column_name", ascending=True)
''')
print()
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ram", "Shyam", "Aman"],
    "Age": [25, 20, 30],
    "Salary": [30000, 20000, 40000]
})

print(df.sort_values(by="Age"))
print()
print(df.sort_values(by="Salary", ascending=False))
print()
print(df)
print(df.sort_values(by=["Salary","Age"], ascending=[True,False],inplace=True))
print()
print("Aggregation----------------------")
print(df["Age"].mean())
print(df["Age"].sum())
print(df["Age"].max())
print(df["Age"].min())
print(df["Age"].count())



```

```python
print("Merging and Joining or concanating")
print()
print("Merging is the process of combining two or more DataFrames based on common column(s), similar to SQL joins.")
print()
print("Joining is the process of combining DataFrames based on their index.")

'''
Syntax--        pd.merge(df1,df2,on="Column_Name",how="type of join")
          join= inner , outer ,left , right, fulljoin
'''
import pandas as pd

df_customers=pd.DataFrame({
    "Name":["Pranjal","Deva","Sam"],
    "Id":[1,2,3]
})
df_orders=pd.DataFrame({
    "Id":[1,2,4],
    "orders":[200,300,40]
})

res=pd.merge(df_customers,df_orders,on="Id",how="inner")
print()

res1=pd.merge(df_customers,df_orders,on="Id",how="outer")
print()

res2=pd.merge(df_customers,df_orders,on="Id",how="left")
print()

res3=pd.merge(df_customers,df_orders,on="Id",how="right")

res4=pd.merge(df_customers,df_orders,how="cross")
print("Inner")
print(res)
print()
print("Outer")
print(res1)
print()
print("left")
print(res2)
print()
print("Right")
print(res3)
print()
print("cross")
print(res4)
print()


print()
df1 = pd.DataFrame({
    'id': [1, 2, 4],
    'name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'id': [1, 2, 4],
    'marks': [90, 80, 70]
})

result = pd.merge(df1, df2, on='id', how='inner')
print(result)

```

```python
print("Concat")
"""
Syntx

pd.concat([df1,df2],axis=0,ignore_index=True)
"""

import pandas as pd

df_Region1=pd.DataFrame({
    'CustomerID':[1,2],
    'Name':["Gopal","Raju"]
})
df_Region2=pd.DataFrame({
    'CustomerID':[3,4],
    'Name':["Gmd","kjen"]
})
print(pd.concat([df_Region1,df_Region2],axis=0,ignore_index=True))
print()
print(pd.concat([df_Region1,df_Region2],axis=1,ignore_index=True))
print()

df1=pd.DataFrame({
    'CustomerID':[1,2,3],
    'Name':["Gopal","Raju","ejnd"]
})
df2=pd.DataFrame({
    'CustomerID':[1,2,4],
    'Transaction':[100,200,300]
})

merged=pd.merge(df1,df2,on="CustomerID",how="outer")
print(merged)
print()

new_customer=pd.DataFrame({
    "CustomerID":[5],
    "Name":["Pranjal"],
    "Transaction":[400]

})

finaldf=pd.concat([merged,new_customer],axis=0,ignore_index=False)
print(finaldf)
print()
finaldf.to_csv("customers.csv",index=False)
```
