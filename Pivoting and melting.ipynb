```python
###pivot 

import pandas as pd
data = {
    "Keys":["k1","k2","k1","k2","k1","k2"],
    "Names": ["HarryPotter", "Hermion", "Ron", "Draco", "Luna", "Nevill"],
    "Houses": ["Gryffindor", "Gryffindor", "Gryffindor", "Slytherin", "Ravenclaw", "Gryffindor"],
    "Grade": [20,30,40,50,40,50]
}

df=pd.DataFrame(data)
# print(df)

pivot_df = df.pivot(index="Keys", columns="Names", values=["Houses","Grade"])
print(pivot_df)
```

```python
import pandas as pd

data = {
    "Names": ["HarryPotter", "Hermion", "Ron", "Draco", "Luna", "Nevill"],
    "Houses": ["Gryffindor", "Gryffindor", "Gryffindor", "Slytherin", "Ravenclaw", "Gryffindor"],
    "Grade": [20, 30, 40, 50, 40, 50]
}

df = pd.DataFrame(data)

# Melt the DataFrame
melt = pd.melt(df, id_vars="Names", value_vars=["Houses"])

print(melt)

# take two parameters

melt1 = pd.melt(df, id_vars="Names", value_vars=["Houses","Grade"],value_name="Values")
print(melt1)


# change variable name


melt2 = pd.melt(df, id_vars="Names", value_vars=["Houses","Grade"],value_name="Values",var_name="Houses&Grade")
print(melt2)


# id_vars---- Specifies the columns to keep fixed ("Names" in this case).
melt3 = pd.melt(df, id_vars=["Names", "Grade"], value_vars=["Houses"], value_name="values")

print(melt3)
```
