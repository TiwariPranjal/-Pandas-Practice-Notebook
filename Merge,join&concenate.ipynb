```python
import pandas as pd
data1 = [
    {"Employee_ID": "E001", "Age": 25, "Name": "Alice"},
    {"Employee_ID": "E002", "Age": 30, "Name": "Bob"},
    {"Employee_ID": "E003", "Age": 28, "Name": "Charlie"},
    {"Employee_ID": "E004", "Age": 35, "Name": "Diana"},
    {"Employee_ID": "E005", "Age": 40, "Name": "Edward"},
]
data2 = [
    {"Employee_ID": "E001", "Salary": 50000, "Department": "HR"},
    {"Employee_ID": "E002", "Salary": 60000, "Department": "IT"},
    {"Employee_ID": "E003", "Salary": 55000, "Department": "Finance"},
    {"Employee_ID": "E004", "Salary": 65000, "Department": "Marketing"},
    {"Employee_ID": "E005", "Salary": 70000, "Department": "Sales"},
    {"Employee_ID": "E006", "Salary": 48000, "Department": "HR"},
    {"Employee_ID": "E007", "Salary": 62000, "Department": "IT"},
    {"Employee_ID": "E008", "Salary": 59000, "Department": "Finance"},
]


df=pd.DataFrame(data1)
print(df)
print()
df1=pd.DataFrame(data2)
print(df1)
print()
print(pd.merge(df,df1,on="Employee_ID"))
print()

print(pd.merge(df,df1,on="Employee_ID",how="inner"))
print()
print(pd.merge(df,df1,on="Employee_ID",how="left"))
print()
print(pd.merge(df,df1,on="Employee_ID",how="right"))
print()
print(pd.merge(df,df1))      #if two data have two different rows ,it take on parameter automatically merge the data in base of any same column

```
